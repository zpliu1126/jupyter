#### 在figure的任意位置添加axes

+ `add_axes()`函数，接收四个参数的列表分别对应`left`、`bottom`、`width`、`high`

如下所示第一幅图在最左边和最下边开始绘制，并且保持宽度和高度占据figure的一半；而第二张图片从figure的中心处开始绘制


````python
import matplotlib.pyplot as plt
import numpy as np

t=np.arange(0,1,0.01)
s=np.sin(2*np.pi*t)

fig1=plt.figure()
axes=fig1.add_axes([0,0,0.5,0.5])  ###第一幅图
axes.plot(t,s,c='b',lw=3)

axes=fig1.add_axes([0.5,0.5,0.5,0.5])  ###第二幅图
line,=axes.plot(t,s,c='b',lw=3)
````

#### 自定义对象
> figure中的每个元素都是通过artist进行渲染的；于此同时figure与axes都是一个rectangle实例对象；通过调用对应的方法从而修改figure、axes的背景色、大小、透明度等其他属性；对应的实力对象存储在`fugure.patch`和`axes.patch`中。

+ `alpha`设置图片透明度
+ `zorder`设置图片显示的顺序

```python
###单个属性的设置方法

###多个属性同时设置的方法
```
